````python
# 모듈 사용
import folium
import json
from folium import Marker

folium.__version__
````

````python
import folium
import json
from folium import Marker

# 서울 중심 위치 설정 (경복궁) - 소수점 4자리 이상 사용해야 정확함
loc = (37.5776087830657, 126.976896737645)

# GeoJSON 파일 열기
json_data = open('./seoul_municipalities_geo_simple.json', encoding='utf-8').read()
data = json.loads(json_data)

# 좌표 데이터 추출
json_coords = []
for i in range(len(data['features'])):
    coords = data['features'][i]['geometry']['coordinates']
    json_coords.append(coords[0][0])  # 다각형의 첫 꼭짓점 좌표 리스트 사용 (외곽선)

# 위도, 경도 분리 저장 (GeoJSON은 [경도, 위도] 순서임)
jsonlat = []  # 위도 리스트
jsonlng = []  # 경도 리스트
for coord in json_coords:
    jsonlng.append(coord[0])  # 경도
    jsonlat.append(coord[1])  # 위도

# 지도 생성
m = folium.Map(location=loc, zoom_start=11)
folium.GeoJson(data, name='동').add_to(m)

# 경복궁 마커 추가
Marker(
    location=loc,
    popup='경복궁',
    icon=folium.Icon(color='lightblue', icon='heart')
).add_to(m)

# folium 아이콘 종류 참고: https://icons.getbootstrap.com/

# 결과 출력 (주피터 노트북 기준)
m
````
